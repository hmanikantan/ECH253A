Here we visualize the fluid velocity field and pressure field associate with the low-Reynolds number motion of a sphere in a viscous fluid. Equivalently, this is the problem of fluid of uniform velocity $U$ flowing past a sphere. The axisymmetric flow field in $(r,\theta)$ works out to be:

$$ u_r(r,\theta)= U \cos\theta \left[ 1- \frac{3}{2}\left( \frac{R}{r} \right) +\frac{1}{2}\left( \frac{R}{r} \right)^3 \right], \\
u_\theta(r,\theta)= - U \sin\theta \left[ 1- \frac{3}{4}\left( \frac{R}{r} \right) -\frac{1}{4}\left( \frac{R}{r} \right)^3 \right],
$$

where $R$ is the radius of the sphere. We can see that the velocities vanish at the sphere surface $r=R$ and match the uniform velocity $U$ far from the sphere, which corresponds to components $(U\cos\theta,-U\sin\theta)$ as $r\rightarrow \infty$ in spherical coordinates.

We can set up and plot the velocity vector quivers in a region around the sphere:

```python
import numpy as np
import matplotlib.pyplot as plt

# take radius to be 1, and lengths nondimensionalized with R
x=np.linspace(-3,3,101)
y=np.linspace(-3,3,101)

X,Y=np.meshgrid(x,y)

R=np.sqrt(X**2+Y**2)
T=np.arctan2(X,Y)

R[R<1]=1

U=1

Ur=U*np.cos(T)*(1-3/2/R+1/2/R**3)
Ut=-U*np.sin(T)*(1-3/4/R-1/4/R**3)

Uy=Ur*np.sin(T)+Ut*np.cos(T)
Uz=Ur*np.cos(T)-Ut*np.sin(T)

Uz[R<1]=0
Uy[R<1]=0

Umag=np.sqrt(Uz**2+Uy**2)
```

```python
figure, axes = plt.subplots(figsize=(10,8))
draw_circle = plt.Circle((0, 0), 1,color='w')

axes.set_aspect(1)
ctr=axes.contourf(x,y,Umag,100,cmap='coolwarm')
s=6 #spacer
axes.quiver(X[::s,::s],Y[::s,::s],Uy[::s,::s],Uz[::s,::s])
axes.add_artist(draw_circle)
figure.colorbar(ctr,ax=axes,fraction=0.1,pad=0.1)
plt.title('velocity vectors of flow around a sphere')
plt.show()
```

The color indicates the magnitude of the velocity: close to the background velocity far away, and equal to zero at the sphere surface.

Streamlines are another way to visualize the flow field:

```python
figure, axes = plt.subplots(figsize=(10,8))
draw_circle = plt.Circle((0, 0), 1,color='w')

axes.set_aspect(1)
ctr=axes.contourf(x,y,Umag,100,cmap='coolwarm')
axes.streamplot(X,Y,Uy,Uz,linewidth=3*Umag,color='k')
axes.add_artist(draw_circle)
axes.set_xlim(-3,3)
figure.colorbar(ctr,ax=axes,fraction=0.1,pad=0.1)
plt.title('streamlines of flow around a sphere')
plt.show()
```

The same plot as above can be obtained from the stream function $\psi(r,\theta)$: lines of constant $\psi$ are indeed the streamlines. Recall that

$$ \psi(r,\theta)= U R^2 \sin^2\theta \left[ \frac{1}{2}\left( \frac{r}{R} \right)^2 -\frac{3}{4}\left( \frac{r}{R} \right)+\frac{1}{4}\left( \frac{R}{r} \right) \right]$$

Plotting contours of this stream function gives:

```python
psi=U*np.sin(T)**2*(1/2*R**2-3/4*R+1/4/R)

figure, axes = plt.subplots(figsize=(10,8))
draw_circle = plt.Circle((0, 0), 1,color='k')

axes.set_aspect(1)
# choice of constant stream function values to pick streamline
str=[0.002,0.02,0.05,0.1,0.2,0.3,0.5,0.7,1,1.5,2]
ctr=axes.contour(x,y,psi,str,cmap='gray')
axes.add_artist(draw_circle)
axes.set_xlim(-3,3)
plt.title('streamlines as curves of constant stream function')
plt.show()
```

To visualize the case of a particle moving in a constant fluid, we can simply add a relative velocity of magnitude $U$ vertically downward to the entire velocity field:

```python
# stagnant fluid, moving particle

Uy=Ur*np.sin(T)+Ut*np.cos(T)
Uz=Ur*np.cos(T)-Ut*np.sin(T)-U

Uz[R<1]=-1
Uy[R<1]=-1

Umag=np.sqrt(Uz**2+Uy**2)

figure, axes = plt.subplots(figsize=(10,8))
draw_circle = plt.Circle((0, 0), 1,color='w')


axes.set_aspect(1)
ctr=axes.contourf(x,y,Umag,100,cmap='coolwarm')
s=6 #spacer
axes.add_artist(draw_circle)
axes.quiver(X[::s,::s],Y[::s,::s],Uy[::s,::s],Uz[::s,::s])
figure.colorbar(ctr,ax=axes,fraction=0.1,pad=0.1)
plt.title('sphere translating downward')
plt.show()
```

Finally, the pressure field works out to be:

$$p(r,\theta) = -\frac{3}{2}\frac{\mu U}{R} \left(\frac{R}{r} \right)^2  \cos\theta+p_\infty$$

Plotting this shows that the pressure field is dipolar: the pressure increases in the front of the sphere relative to the fluid flow, and drops in the rear.

```python
# pressure field

p=-1.5/R**2*np.cos(T)

figure, axes = plt.subplots(figsize=(10,8))
draw_circle = plt.Circle((0, 0), 1,color='w')

axes.set_aspect(1)
ctr=axes.contourf(x,y,p,100,cmap='coolwarm')
axes.add_artist(draw_circle)
figure.colorbar(ctr,ax=axes,fraction=0.1,pad=0.1)
plt.title('pressure field')
plt.show()
```
